```python
import librosa
audio_path = r"C:\Users\DELL\Desktop\AVV_DATASET\Male\SAD\DC_sa07.wav"
x, sr = librosa.load(audio_path)
```

```python
librosa.load(audio_path, sr=44100)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 5))
plt.plot(x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Waveform')
plt.show()
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
print(Xdb)
```

```python
n0 = 9000
n1 = 9100
plt.figure(figsize=(14, 5))
plt.plot(x[n0:n1])
plt.grid()
```

```python
zero_crossings = librosa.zero_crossings(x[n0:n1], pad=False)
print(sum(zero_crossings))
```

```python
import sklearn.preprocessing
import numpy as np

# Calculate spectral centroid
spectral_centroids = librosa.feature.spectral_centroid(y=x, sr=sr)[0]

# Computing the time variable for visualization
frames = range(len(spectral_centroids))
t = librosa.frames_to_time(frames)

# Normalizing the spectral centroid for visualization
def normalize(x, axis=0):
    return sklearn.preprocessing.minmax_scale(x, axis=axis)

# Plotting the waveform and spectral centroid
plt.figure(figsize=(14, 5))
plt.plot(np.linspace(0, len(x) / sr, num=len(x)), x, alpha=0.4, label='Waveform')
plt.plot(t, normalize(spectral_centroids), color='r', label='Spectral Centroid')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude / Spectral Centroid')
plt.title('Waveform with Spectral Centroid')
plt.legend()
plt.show()
```

```python
mfccs = librosa.feature.mfcc(y = x, sr = sr)
print(mfccs.shape)
#Displaying  the MFCCs:
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
```

```python
tonal_centroid = librosa.feature.tonnetz(y=x, sr=sr)
print(tonal_centroid.shape)
plt.figure(figsize=(10, 6))
plt.imshow(tonal_centroid, aspect='auto', origin='lower', cmap='jet')
plt.colorbar()
plt.title('Tonal Centroid')
plt.xlabel('Time')
plt.ylabel('Tonal Centroid')
plt.tight_layout()
plt.show()
```

```python
harmonic, percussive = librosa.effects.hpss(x)

# Compute the fundamental frequency (pitch)
f0, voiced_flag, voiced_probs = librosa.pyin(y=harmonic, fmin=librosa.note_to_hz('C1'), fmax=librosa.note_to_hz('C7'))
plt.figure(figsize=(10, 6))
plt.plot(librosa.times_like(f0), f0, label='F0 (fundamental frequency)')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.title('Fundamental Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mfccs = librosa.feature.mfcc(y=x, sr=sr, n_mfcc=13)  # You may adjust n_mfcc as needed

# Extract the 0th coefficient
zeroth_mfcc = mfccs[0]

# Plot the 0th coefficient
plt.figure(figsize=(10, 6))
plt.plot(librosa.times_like(zeroth_mfcc), zeroth_mfcc, label='0th Cepstral Coefficient')
plt.xlabel('Time (s)')
plt.ylabel('Coefficient Value')
plt.title('0th Cepstral Coefficient')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
spectral_contrast = librosa.feature.spectral_contrast(y=x, sr=sr)

# Plot spectral contrast
plt.figure(figsize=(10, 6))
librosa.display.specshow(spectral_contrast, sr=sr, x_axis='time')
plt.colorbar()
plt.title('Spectral Contrast')
plt.ylabel('Frequency Bands')
plt.xlabel('Time')
plt.tight_layout()
plt.show()
```

```python
stft = librosa.stft(x)

# Compute the chromagram
chromagram = librosa.feature.chroma_stft(S=stft, sr=sr)

# Plot the chromagram
plt.figure(figsize=(10, 6))
librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', cmap='coolwarm')
plt.colorbar()
plt.title('Chromagram')
plt.xlabel('Time (s)')
plt.ylabel('Pitch Class')
plt.tight_layout()
plt.show()
```

```python
stft = librosa.stft(x)

# Compute the chromagram
chromagram = librosa.feature.chroma_stft(S=stft, sr=sr)

# Compute delta coefficients of the chromagram
delta_chromagram = librosa.feature.delta(chromagram)

# Plot the chromagram
# plt.figure(figsize=(10, 6))
# plt.subplot(2, 1, 1)
# librosa.display.specshow(chromagram, x_axis='time', y_axis='chroma', cmap='coolwarm')
# plt.colorbar()
# plt.title('Chromagram')
# plt.xlabel('Time (s)')
# plt.ylabel('Pitch Class')

# Plot the delta coefficients
plt.figure(figsize=(10, 6))
librosa.display.specshow(delta_chromagram, x_axis='time', y_axis='chroma', cmap='coolwarm')
plt.colorbar()
plt.title('Delta Chromagram')
plt.xlabel('Time (s)')
plt.ylabel('Pitch Class')

plt.tight_layout()
plt.show()
```

```python
frame_size = 1024
hop_length = 512

# Compute short-term energy
energy = librosa.feature.rms(y=x, frame_length=frame_size, hop_length=hop_length)[0]

# Compute time axis for plotting
time = librosa.times_like(energy, sr=sr, hop_length=hop_length)

# Plot short-term energy
plt.figure(figsize=(10, 6))
plt.plot(time, energy, color='b', label='Short-term Energy')
plt.xlabel('Time (s)')
plt.ylabel('Energy')
plt.title('Short-term Energy of Speech Signal')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
